```python
import pandas as pd
```

```python
data_dir = 'C:/dev/git_jupyter/jupyter_repo/data/'

df_tr_master = pd.read_excel(data_dir+'transformer_data.xlsx', sheetname='master')
df_tr_master
```

```python
df_tr_wo = pd.read_excel(data_dir+'transformer_data.xlsx', sheetname='workorder')
df_tr_wo
```

```python
df_tr_all = pd.merge(df_tr_master, df_tr_wo)
df_tr_all
```

```python
# df_tr_all['delta'] = (df_tr_all['RepairDate']-df_tr_all['RepairDate'].shift()).fillna(0)
# df_tr_all.reset_index()
df_tr_all['RepairDate'] = pd.to_datetime(df_tr_all['RepairDate'], dayfirst=True)
df_tr_all
```
